```python
import cv2
import matplotlib.pyplot as plt
input = cv2.imread('tamater.png')
plt.imshow(input)
plt.show()
```

```python
plt.hist(input.ravel(), bins = 256, range = [0, 256])
plt.show()
```

```python
color = ('b', 'g', 'r')
for i, col in enumerate(color):
    histr = cv2.calcHist([input],[i],None,[256],[0, 256])
    plt.plot(histr, color = col)
    plt.xlim([0, 256])
    plt.show()
```

```python
input = cv2.cvtColor(input,cv2.COLOR_BGR2GRAY)
plt.imshow(input)
plt.show()
```

```python
plt.hist(input.ravel(), bins = 256, range = [0, 256])
plt.show()
```

```python
im2 = 255 - (input)
plt.imshow(im2)
plt.show()
```
